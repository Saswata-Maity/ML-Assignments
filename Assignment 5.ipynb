```python
df=df.drop(['loan_id'],axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df.columns=df.columns.str.strip()
```

```python
df['loan_status']=le.fit_transform(df['loan_status'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x.head(2)
```

```python
y.head(2)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
from sklearn.linear_model  import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
lr=LogisticRegression(max_iter=1000)#wasnt converging on the default 100 iterations
sv=SVC()
rf=RandomForestClassifier()
knn=KNeighborsClassifier()
```

```python
lr.fit(x_train,y_train)
```

```python
sv.fit(x_train,y_train)
```

```python
rf.fit(x_train,y_train)
```

```python
knn.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score,accuracy_score,classification_report
```

```python
ypred_l=lr.predict(x_test)
ypred_s=sv.predict(x_test)
y_pred_rf=rf.predict(x_test)
y_pred_kn=knn.predict(x_test)
```

```python
d={"Algorithms":pd.Series(["Logistic Regression","Support Vector Machine","Random Forest","Knn Classifer"]),
    "Accuracy Scores": pd.Series([accuracy_score(y_test,ypred_l),accuracy_score(y_test,ypred_s),accuracy_score(y_test,y_pred_rf),accuracy_score(y_test,y_pred_kn)])
}
```

```python
print(pd.DataFrame(data=d))
```

```python
print("Logistic Regression")
print(classification_report(y_test,ypred_l))
```

```python
print("Support Vector Machine")
print(classification_report(y_test,ypred_s))
```

```python
print("RandomForest")
print(classification_report(y_test,y_pred_rf))
```

```python
print("Knearest Neighbors")
print(classification_report(y_test,y_pred_kn))
```
